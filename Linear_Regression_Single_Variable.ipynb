```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = {
    "area": [2600, 3000, 3200, 3600, 4000],
    "price": [550000, 565000, 610000, 680000, 725000]
}

df = pd.DataFrame(data)
print(df)
```

```python
%matplotlib inline
plt.scatter(df.area, df.price, color="red", marker="+")
plt.xlabel("Area")
plt.ylabel("Price")
```

```python
print(df.price)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["area"]], df.price)
```

```python
print("Slope (coef):", reg.coef_)
print("Intercept:", reg.intercept_)
```

```python
reg.predict([[2600]])
```

```python
data = {
    "area": [2400,1000,1500, 3200,465,3433,7888]
}
d=pd.DataFrame(data)
```

```python
p=reg.predict(d)
p
```

```python
d["prices"]=p
```
